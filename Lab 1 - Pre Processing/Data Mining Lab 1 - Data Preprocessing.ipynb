# 16. Come up with a new feature - All star users
### these users are ones that have given at least 1 review from each star category:
### 1x 1-star review, 1x 2-star review, 1x 3-star review, 1x 4-star review, 1x 5-star review
### This feature is a binary variable - either you have reviewed all of these categories or you haven't

```python
users_stars = reviews.groupby('user_id')
f = lambda x: len(set(x))
b = users_stars['stars'].apply(f)
b_df = pd.DataFrame(b)
b_df["all_star"] = b_df[b_df['stars']==5]
```

```python
users_all_stars_df = b_df["all_star"]
```

```python
users_all_stars_df = pd.DataFrame(users_all_stars_df)
users_all_stars_df
```

```python
final_answer = users_all_stars_df.applymap(lambda x: 1 if x==5 else 0)
final_answer['user_id']=final_answer.index
final_answer['all_stars'] = final_answer['all_star']
del final_answer['all_star'] 
del final_answer['user_id']
final_answer
```
